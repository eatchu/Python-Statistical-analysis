# 1. DF Merge

- DF1 + DF2 -> DF3
- 사용함수 : concat, merge

Series merge : 1차원 병합
 - 형식) pandas.concat([data1,data2,..], axis = 0 or 1)

```python
from pandas import Series
s1 = Series([1, 3], index=['a','b'])
s2 = Series([5, 6, 7], index=['a','b','c'])
s3 = Series([11, 13], index=['a','b'])

# 행 단위 결합 : rbind()
import pandas as pd
s4 = pd.concat([s1,s2,s3], axis=0)
print('**********열단위 결합**********')
print(s4)
s4.shape # (7,)

# 열 단위 결합
s5 = pd.concat([s1,s2,s3], axis=1)
print('**********행단위 결합**********')
print(s5)
s5.shape # (3,3)
```

DataFrame merge : 2차원 병합
 - 형식1) pandas.merge(DF1, DF2,...) : 공통칼럼 존재해야함
 - 형식2) pandas.concat([DF1,DF2,...], axis = 0 or 1)

```python
wdbc = pd.read_csv('C:\\IITT\\4_Python-II\\workplace\\data\\wdbc_data.csv', encoding='utf-8')
wdbc.info() # 569 x 32

# DF1(16) + DF2(16) : slicing
cols = list(wdbc.columns)

DF1 = wdbc[cols[:16]]
DF2 = wdbc[cols[16:]]


# id 칼럼 추가 : 공통 칼럼 생성을 위해
id = wdbc.id
DF2['id'] = id 


# DF 병합 : 공통 칼럼 존재해야함
DF3 = pd.merge(DF1, DF2)
print('********** merge 함수 사용**********')
print(DF3) # 569 x 32


# 칼럼 병합 : 공통 칼럼 존재하지 않아도 됨
DF4 = pd.concat([DF1,DF2],axis=1) # id 동일칼럼이 두개 생김
print('********** concat 함수 사용**********')
print(DF4) # 569 x 33
```

# 2. Reshape
- 형식) wide -> long : stack()
- 형식) long -> wide : unstack()

```python

import pandas as pd

buy = pd.read_csv('C:\\IITT\\4_Python-II\\workplace\\data\\buy_data.csv', encoding='utf-8')
buy.info() # 22 x 3


# 1. row -> column (wide -> long)
buy_long = buy.stack() # 1차원 vector 형태로 만듬 : 22x3 -> 1x66
print('********** wide -> long **********')
print(buy_long)

# 2. column -> row (long -> wide)
buy_wide = buy_long.unstack()
print('********** long -> wide **********')
print(buy_wide) # 원상태로 되돌아옴 2차원 df : 1x66 -> 22x3

# 3. 전치 행렬 column,row -> row,column
wide_t = buy_wide.T
print('********** 전치행렬1 **********')
print(wide_t) # 22x3 -> 3x22
long_t = wide_t.T
print('********** 전치행렬2 **********')
print(long_t) # 3x22 -> 22x3

# 4. 중복 행 제거 
buy_df = buy.drop_duplicates()
print('********** 중복행 제거 **********')
print(buy_df.shape) # 20x3 : 2개의 중복행 제거
```

# 3. csv file
- csv file read ) pandas.read_csv('경로/csv')
- csv file save ) obj.to_csv('경로/csv')

```python
import pandas as pd

# 1) csv file read
iris = pd.read_csv('C:\\IITT\\4_Python-II\\workplace\\data\\iris.csv')
print('********** read csv **********')
print(iris.info())


# 칼럼명 : 특수문자 or 공백 -> _문자변경
iris.columns = iris.columns.str.replace('.','_')
# Sepal.Width -> Sepal_Width


# 칼럼명이 없는 경우
st = pd.read_csv('C:\\IITT\\4_Python-II\\workplace\\data\\student.csv', header=None)
print('********** 칼럼명이 없는 csv file read **********')
print(st)

# 비만도 지수(BMI) 칼럼 추가
# BMI = 몸무게/키**2 (몸무게:kg, 키:m)
BMI = [st.loc[i,3]/(st.loc[i,2]*0.01)**2 
       for i in range(len(st))]
st['bmi'] = BMI
st['bmi'] = st['bmi'].round(2)



# 2) csv file save
st.to_csv('C:\\IITT\\4_Python-II\\workplace\\data\\student_df.csv', index=None, encoding='utf-8')


# save file 확인
st_df = pd.read_csv('C:\\IITT\\4_Python-II\\workplace\\data\\student_df.csv')
print('********** save file 확인 **********')
print(st_df)



```
