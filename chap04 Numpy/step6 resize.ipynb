```python

"""
reshape vs resize
 - reshape : 모양 변경
 - resize : 크기 변경
   ex) images -> 120x150 규격화 -> model

image 규격화 : 실습
"""

from glob import glob # file 검색 패턴 사용 (문자열 경로, *.jpg)
from PIL import Image # image file read
import numpy as np
import matplotlib.pylab as plt


# 1개 image file open
file = 'C://IITT/4_Python-II/workplace/chap04_Numpy/images/test1.jpg'
img = Image.open(file) # image file read
type(img) # PIL.JpegImagePlugin.JpegImageFile
np.shape(img) # (360,540,3)
plt.imshow(img)


img_re = img.resize((150,120)) # w,h
np.shape(img_re) # (120,150,3)
plt.imshow(img_re)


# PIL -> numpy : image handling
type(img_re) # PIL.Image.Image
type(img) # PIL.JpegImagePlugin.JpegImageFile
img_arr = np.asarray(img)
type(img_arr) # numpy.ndarray
img_arr.shape # (360, 540, 3)




# 여러장의 image resize 함수
def imageResize():
    img_h = 120 # 가로픽셀
    img_w = 150 # 세로픽셀
    
    image_resize = [] # 규격화된 image 저장
    
    # glab : file 패턴
    for rile in glob('C:/IITT/4_Python-II/workplace/chap04_Numpy/images/'+'*.jpg'):
        # test1.jpg -> test2.jpg, ...
        img = Image.open(file) # image file read
        print(np.shape(img)) # img shape
        
        # PIL -> resize
        img = img.resize((img_w, img_h)) # w,h
        # PIL -> numpy
        img_data = np.asarray(img)
        
        # resize image save
        image_resize.append(img_data)
        print(file, ':', img_data.shape) # image shape
        
    return image_resize   


image_resize = imageResize()
'''
(360, 540, 3)
chap04_Numpy/images/test1.jpg : (120, 150, 3)
(360, 540, 3)
chap04_Numpy/images/test1.jpg : (120, 150, 3)
'''




image_resize[0].shape # (120,150,3)
```
