       [3,4] -> [4,2]
      

```python

import numpy as np
from matplotlib.image import imread # image 읽기
import matplotlib.pylab as plt



# 1. image read
file_path = 'C:/IITT/4_Python-II/workplace/chap04_Numpy/images/test1.jpg'
image = imread(file_path)
type(image) # numpy.ndarray

image.shape # (360, 540, 3) -> (가로, 세로, 컬러)
print(image)

plt.imshow(image) # 이미지 보기


# RGB 색상 분류
r = image[:,:,0] # R
g = image[:,:,1] # G
b = image[:,:,2] # B
r.shape # (360, 540)
g.shape # (360, 540)
b.shape # (360, 540)




# 2. image data reshape
from sklearn.datasets import load_digits

digit = load_digits() # dataset loading
digit.DESCR # 설명보기

X = digit.data # x변수(입력변수) : image
Y = digit.target # y변수(정답=정수)

X.shape # (1797,64) 64=8x8
Y.shape # (1797,)

img_0 = X[0].reshape(8,8)
img_0
plt.imshow(img_0) # 0
Y[0] # 0


X_3d = X.reshape(-1,8,8)
X_3d.shape # (1797,8,8) 3차원 -> (전체이미지, 세로, 가로, [칼럼])

# (1797,8,8,1)
X_4d = X_3d[:,:,:,np.newaxis] # 4번째에 새로운 축 추가
X_4d.shape # (1797, 8, 8, 1)






# 3. data reshape
'''
T : 전치행렬1
swapaxis : 전치행렬2
transpose() : 3차원 이상 모양 변경
'''

# 1) 전치행렬
data = np.arange(10).reshape(2,5)
print(data)
print(data.T)


# 2) transpose()
'''
1차원 : 효과 없음
2차원 : 전치행렬 동일
3차원 : (0,1,2) -> (2,1,0)
'''
arr3d = np.arange(1,25).reshape(4,2,3) # 4(2x3)
arr3d.shape # (4,2,3)
arr3d


# (0,1,2) -> (2,1,0)
arr3d_tran = arr3d.transpose()
arr3d_tran.shape # (3,2,4)
arr3d_tran

# (0,1,2) -> (1,2,0)
arr3d_tran2 = arr3d.transpose(1,2,0)
arr3d_tran2.shape # (2,3,4)
arr3d_tran2

```
