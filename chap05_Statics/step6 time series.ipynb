# 시계열분석 (time series analysis)
1. 시계열 자료 생성
2. 날짜형 수정 (다국어)
3. 시계열 시각화
4. 이동평균 기능 : 5,10,20일 평균 -> 추세선 평활 (스뮤딩)

```python
# 패키지 설치
from datetime import datetime # 날짜형식 변경
import pandas as pd # csv file read
import matplotlib.pyplot as plt # 시계열 시각화
import numpy as np # 수치 자료 생성
```

# 1. 시계열 자료생성

```python
# 일(day)단위 데이터 생성 : freq='D'(default)
time_data = pd.date_range("2017-03-01","2020-03-30")
time_data # len=1126, freq="D" 

# 월(month)단위 데이터 생성 : freq='M'
time_data2 = pd.date_range("2017-03-01","2020-03-30",freq='M')
time_data2 # freq='M
len(time_data2) # 36

# 월 단위 매출 현황
x = pd.Series(np.random.uniform(10,100,size=36))
df = pd.DataFrame({'data':time_data2,'price':x})
df

# 시각화
plt.plot(df['data'],df['price'],'g') # (x,y)
plt.title('month-price')
plt.show()
```

# 2. 날짜형식 수정 (다국어)

```python
cospi = pd.read_csv('C:/IITT/4_Python-II/workplace/data/cospi.csv')
# cospi.info() # 247 x 6
print(cospi.head()) # Date형식) day-month-year : 다국어 형식

date = cospi['Date']
len(date) # 247
```

```python
'''
날짜 및 시간 지정 문자열	의미
%Y	앞의 빈자리를 0으로 채우는 4자리 연도 숫자
%m	앞의 빈자리를 0으로 채우는 2자리 월 숫자
%d	앞의 빈자리를 0으로 채우는 2자리 일 숫자
%H	앞의 빈자리를 0으로 채우는 24시간 형식 2자리 시간 숫자
%M	앞의 빈자리를 0으로 채우는 2자리 분 숫자
%S	앞의 빈자리를 0으로 채우는 2자리 초 숫자
%A	영어로 된 요일 문자열
%B	영어로 된 월 문자열
'''

kdate = [datetime.strptime(d,'%d-%b-%y') for d in date] # strftime
 
# 날짜 칼럼 수정 
cospi['Date'] = kdate
print(cospi.head()) # 26-Feb-16 -> 2016-02-26
# cospi.info() # object -> datetime
```

# 3. 시계열 시각화

```python
cospi.index # (start=0, stop=247, step=1)

# 칼럼 -> index 적용 : 숫자 index를 날짜 index로 변경해줌
new_cospi = cospi.set_index('Date')
new_cospi.index # index -> Date

new_cospi['2016'] # 2016년도에 대한 데이터 추출

new_cospi['2015-05':'2015-03'] # 월별 접근 가능

# subset
cospi_HL = new_cospi[['High','Low']]
cospi_HL.index # Date
cospi_HL.columns # High, Low

# 2015년 기준
cospi_HL['2015'].plot(title='2015year : High vs Low')
plt.show()

# 2016년 기준
cospi_HL['2016'].plot(title='2016year : High vs Low')
plt.show()

# 2016년 2월 기준
cospi_HL['2016-02'].plot(title='2016year : High vs Low')
plt.show()
```

# 4. 이동평균 기능 : 5, 10, 20일 평균 -> 추세선 평활(스뮤딩)

```python
# 1) 5일 단위 이동평균 : 5일 단위 평균 -> 마지막 5일째 이동
roll_mean5 = pd.Series.rolling(cospi_HL, window=5, center=False).mean()
print(roll_mean5.head())

# 2) 10일 단위 이동평균 : 5일 단위 평균 -> 마지막 5일째 이동
roll_mean10 = pd.Series.rolling(cospi_HL, window=10, center=False).mean()
print(roll_mean10.head())

# 3) 20일 단위 이동평균 : 5일 단위 평균 -> 마지막 5일째 이동
roll_mean20 = pd.Series.rolling(cospi_HL, window=20, center=False).mean()
print(roll_mean20.head())


# rolling mean 시각화
cospi_HL.High.plot(color='b',label='High column')
roll_mean5.High.plot(color='r',label='rolling mean 5day')
roll_mean10.High.plot(color='g',label='rolling mean 10day')
roll_mean20.High.plot(color='orange',label='rolling mean 20day')
plt.legend(loc='best')
plt.show()
```
