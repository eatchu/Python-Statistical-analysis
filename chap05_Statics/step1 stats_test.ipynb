# scipy 패키지의 확률분포 검정
 1. 정규분포 검정 : 연속변수의 확률분포
- 연속확률분포 : 정규분포, 균등분포, 카이제곱, T/Z/F분포
 2. 이항분포 검정 : 2가지(성공/실패) 범주의 확률분포
- 이산확률분포 : 베르누이분포, 이항분포, 포아송분포

```python
# 패키지 
from scipy import stats # 확률분포검정
import numpy as np
import matplotlib.pyplot as plt # 시각화
```

# 1. 정규분포 검정 : 평균 중심 좌우 대칭성 검정

```python
# 1) 정규분포객체 생성
mu = 0; std = 1 # 표준정규분포
std_norm = stats.norm(mu,std) # 표준정규분포객체 생성
std_norm # object.info

# 2) 정규분포확률변수
N = 1000 # 시행횟수
norm_data = std_norm.rvs(N) # 시뮬레이션 : 1000난수 생성
len(norm_data) # 1000
norm_data
norm_data.mean() # 0
norm_data.std() # 1

# 3) 히스토그램
plt.hist(norm_data)
plt.show()
plt.plot(norm_data)
plt.show()

# 4) 정규성 검정
# 귀무가설(H0) : 정규분포와 차이가 없다

t_stats, pvalue = stats.shapiro(norm_data)

print('검정통계량:%.5f, pvalue:%.5f'%(t_stats,pvalue))
print('\n','*'*50,'\n')
print('귀무가설 채택 : 정규분포와 차이가 없다.' if pvalue>=0.05 else '귀무가설 기각 : 정규분포와 차이가 있다.')
```

# 2. 이항분포 검정 : 2가지 범주의 확률분포
 - 베르누이분포 : 이항변수(성공 or 실패)에서 성공(1)이 나올 확률분포
 - 이항분포 : 베르누이 분포에 시행횟수(N)을 적용한 확률분포

```python
# example) 게임을 100번의 시행횟수로 정하고 게임에 이길 확률이 40%라면 성공횟수는 몇번일까

N = 100
P = 0.4

# 1) 베르누이 분포 확률변수
x = stats.bernoulli(P).rvs(N)
x # 0(실패) or 1(성공)
len(x) # 100


# 2) 성공횟수
x_succ = np.count_nonzero(x) 
print('성공횟수 = ', x_succ) 


# 3) 이항분포 검정 : 이항분포에 대한 가설검정

# 귀무가설 : 게임에 이길 확률은 40%와 다르지 않다.
pvalue = stats.binom_test(x= x_succ, n = N, p=0.5, alternative='two-sided') # 양측검정
'''
x : 성공횟수
n : 시행횟수
p : 성공확률
alternative='two-sided' : 양측검정(기본이 양측검정)
'''
print('pvalue = ', pvalue) 

if pvalue >= 0.05:
    print('게임에 이길 확률은 40%와 다르지 않다.')
else :
    print('게임에 이길 확률은 40%와 다르다고 볼 수 있다.')
```
