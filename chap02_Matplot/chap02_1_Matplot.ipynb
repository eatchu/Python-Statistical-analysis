matplotlib API 사용해 차트 그리기

- 형식) plt.plot(data) ; plt.show()


# 1. 기본차트 그리기
- plt.plot()

```python
import matplotlib.pyplot as plt 
import pandas as pd # file read
import numpy as np # 숫자 데이터 생성



help(plt.plot)
'''
plot(x, y)        # plot x and y using default line style and color
plot(x, y, 'bo')  # plot x and y using blue circle markers
plot(y)           # plot y using x as index array 0..N-1
plot(y, 'r+')     # ditto, but with red plusses
'''

# 간단한 기본차트
data = np.arange(10) # 0~9
plt.plot(data, 'r') 
print('********** 0~9 일련의 숫자 선차트 ***********')
plt.show()


# 평균=0, 표준편차=1 : 표준정규분포 차트
data2 = np.random.randn(10) # 표준정규분포값 10개를 임의로 생성
plt.plot(data2,'g')
print('********** 랜덤한 실수 선 그래프 **********')
plt.show()

plt.plot(data,data2,'b') # 선 그래프
plt.plot(data,data2,'ro') # 산점도 그래프
print('********** 두개 데이터의 그래프 **********')
plt.show()
```

# 2. 산점도 그리기
- plt.scatter()

```python

# 단색 산점도
plt.scatter(data,data2,c='g',marker='*')
plt.show()

# 여러색 산점도
cdata = np.random.randint(1,4,10) # 1~3의 정수를 임의로 10개 생성
plt.scatter(data, data2, c=cdata, marker='o')ekst
plt.show()
```

# 3. subplot 차트 그리기

```python
fig = plt.figure(figsize = (10,7)) # 차트 크기 (가로:5 세로:3)
x1 = fig.add_subplot(2,2,1)
x2 = fig.add_subplot(2,2,2)
x3 = fig.add_subplot(2,2,3)
x4 = fig.add_subplot(2,2,4)

# data 생성
data3 = np.random.randint(1,100,100)
data4 = np.random.randint(10,110,100)
cdata = np.random.randint(1,4,100)

# 첫번째 격자 : 히스토그램
x1.hist(data3)

# 두번째 격자 : 산점도
x2.scatter(data3,data4, c=cdata)

# 세번째 격자 : 선 그래프
x3.plot(data3)

# 네번째 격자 : 점선 그래프 (선 그래프 옵션 )
x4.plot(data4,'g--')
plt.show()
```

# 4. 차트 크기 지정, 두개 이상 차트 그리기

```python
fig = plt.figure(figsize=(12,7))
chart = fig.add_subplot() # (1,1,1)

# 계단형 차트
chart.plot(data, color='r', label='step', drawstyle='steps-post')

# 선스타일 차트
chart.plot(data2, color='b', label='line')


# 차트에서 한글 지원 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

# 음수 부호 지원 
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False


# 차트 제목
plt.title('계단형 vs 선 스타일')
plt.xlabel('데이터')
plt.ylabel('난수 정수')
plt.legend(loc='best')

plt.show()
```
